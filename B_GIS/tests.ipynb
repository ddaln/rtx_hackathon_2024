```python
df = pd.read_csv('C:/Users/alan1/Documents/Projects/rtx_hackathon_2024/B_GIS/data/Tallest_Towers.csv')
df
```

```python
import folium
sf_map = folium.Map(location=[37.7750, -122.4200], zoom_start=12)

for idx, row in df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        tooltip=row['Name'],
        popup=row['Name'],
        icon=folium.Icon(color='lightgray', icon='home', prefix='fa')
    ).add_to(sf_map)

sf_map.save('data/sf_towers_map.html')

sf_map
```
